```python
# 4. Рассчитаем сколько квартир представлено в каждом районе.
```

```python
hr.Район.unique()
```

```python
hr['Район'].value_counts()
```

```python
# Большая часть квартир представлено во Всеволожском районе ЛО, Приморском и Выборгском районах.
```

```python
dist = round(hr['Район'].value_counts()*100/sum(hr['Район'].value_counts()))
print(f'\nРаспределение доли квартир по районам в процентном соотношении:\n{dist}')
```

```python
# 5. Покажем распределение квартир по ценам.
```

```python
print(hr['ЦенаСкидки'].value_counts())
print('\nРаспределение доли квартир с уровнем цены в процентном соотношении:\n'
      , round(hr['ЦенаСкидки'].value_counts()*100/sum(hr['ЦенаСкидки'].value_counts())))
pic = hr['ЦенаСкидки'].value_counts()
pic.plot.bar()
```

```python
# Большая часть квартир имеет низкий и средний уровни зарплаты - 92% в сумме, с высоким уровнем дохода всего лишь 8%.
```

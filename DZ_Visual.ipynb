```python
df_happy = pd.read_csv('Happines2019.csv')
df_happy.rename(columns={'Country or region': 'Country_or_region',
                         'Social support': 'Social_support',
                         'GDP per capita': 'GDP_per_capita',
                         'Healthy life expectancy': 'Healthy_life_expectancy', 
                         'Perceptions of corruption': 'Perceptions_of_corruption',
                         'Freedom to make life choices': 'Freedom_choices'}, inplace=True)
pd.options.display.float_format = '{:,.3f}'.format
# df_happy.head()
# df_happy.info()
# пример сбора данных для построения графика matplotlib
# happiness = df_happy.groupby('Country_or_region')[['Score', 'Social_support', 'Perceptions_of_corruption']].sum().sort_values('Score', ascending=False)
# happiness = df_happy[[x for x in df_happy.columns if 'Score' in x] + ['Country_or_region']].\
#             groupby('Country_or_region').sum().sort_values('Score', ascending=False)
# happiness
```

```python
ax = sns.distplot(df_happy.Score)
ax.set_title('Распределение баллов по уровню счастья в странах')
```

```python
# Гистограмма показывает распределение баллов по уровню счастья в разных странах. 
# Большинство стран находится в диапазонах 4-6.5 баллов, что говорит о среднем типе удовлетворенности граждан по 
# выбранным государствам. Количество стран с уровнем удовлетворенности выше среднего больше, 
# чем с уровнем ниже среднего. Государства с максимальным уровем удовлетворенности в данной выборке отсутствуют, что говорит 
# о том, что есть потенциал улучшения для всех.
```

```python
sns.pairplot(df_happy[['Score', 'Social_support', 'GDP_per_capita', 'Healthy_life_expectancy', 'Perceptions_of_corruption']])
```

```python
# Чем выше уровень счастья, тем более удовлетворены граждане уровню социальной поддержки, тем выше уровень ВВП на душу населения
# и тем выше продолжительность жизни. Оцениваемый уровень коррупции в 
# странах примерно на одном (низком) уровне, правда видно из графиков, что она оценивается более высоко в странах с самым 
# высоким уровнем счастья (есть исключения, но это единичные случаи из стран с меньшим уровнем удовлетворенности граждан). 
# Это говорит о том, что страны в основном оцениваются как более коррумпированые.
# С малым уровнем коррупции - только страны с наивысшим уровнем счастья. 
# Наибольший удельный вес в оценке уровня счастья в странах оказывают такие два показателя как ВВП на душу населения и 
# уровень социальной поддержки. Уровень коррупции и продолжительность жизни оказывают меньший удельный вес.
```

```python
google = pd.read_csv('googleplaystore.csv')
google.head()
```

```python
google_f1 = google.groupby('Category').count().sort_values('App', ascending=False)
google_f1 = google_f1[['App']]
google_f1.head()
```

```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 13,6

google_f1.loc[['FAMILY', 'GAME', 'TOOLS', 'MEDICAL', 'BUSINESS']].plot(kind='bar', 
                                                                  title='Количество приложений по категориям',
                                                                  xlabel='Категории',
                                                                  ylabel='Количество приложений')
```

```python
# Топ 5 приложений по количеству наименований, доступных на площадке Google Play. 
```

```python
google_f2 = google.groupby('Category').sum().sort_values('Rating', ascending=False)
# google_f = google_f[['App', 'Rating']]
google_f2.head()
```

```python
google_f2.loc[['FAMILY', 'GAME', 'TOOLS', 'PRODUCTIVITY', 'MEDICAL']].plot(kind='bar', 
                                                                  title='Количество приложений по категориям',
                                                                  xlabel='Категории',
                                                                  ylabel='Количество приложений')
```

```python
# Топ 5 приложений по количеству наименований, доступных на площадке Google Play. 
# Первая строчка рейтинга - в категории FAMILY, за ним идет категория GAME, TOOLS, 
# PRODUCTIVITY, MEDICAL. Фиксируется взаимосвязь количества приложений определенной категории 
# и ее суммарного рейтинга. Данный график можно считать неинформативным, так как идет практически полное наследование от 
# предыдущего графика (в помощь категориям приходит их общая сумма рейтинга по количеству).
```
